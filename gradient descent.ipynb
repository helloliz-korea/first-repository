```python
import numpy as np
import matplotlib.pyplot as plt
```

$\min_{x}f(x) = \min_{x} x^{2} - 4x + 6$

```python
f = lambda x: x**2 - 4*x + 6
x = np.linspace(-1, 6, 100)  # domain(정의역)
y = f(x)
```

```python
plt.figure(figsize = (10, 5))
plt.plot(x, y, 'k-')
plt.plot(2, 2, 'sk') # a minimum value
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('$x^{2}-4x+6$')
```

```python
grad_f = lambda x: 2*x - 4 #gradient
```

```python
x0 = 0 
epoch = 10 
learning_rate =  0.25

print("step\t x \t f(x)")
print("{0:02d}\t{1:6.5f}\t{2:6.5f}".format(0, x0, f(x0)))

x_do = [x0]
y_hat = [f(x0)]

for i in range(0, epoch):
    x1 = x0 - learning_rate * grad_f(x0)
    x0 = x1
    print("{0:02d}\t{1:6.5f}\t{2:6.5f}".format(i+1, x0, f(x0)))
    x_do.append(x0)
    y_hat.append(f(x0))
```

```python
plt.figure(figsize = (10, 5))
plt.plot(x, y, 'k-')
plt.plot(2, 2, 'sk') # a minimum value
plt.plot(x_do, y_hat, 'b*:')
for k, point in enumerate(x_do):
    plt.text(point, f(point), '$x_{}$'.format(k), 
            verticalalignment = 'bottom')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('$x^{2}-4x+6$')
```

```python
def steepest_descent(f, grad_f, init, learning_rate = 1e-1, epochs = 10, disp = True):

    x0 = init
    
    if disp == True:
        print("step\t x \t f(x)")
        print("{0:02d}\t{1:>8.5f}\t{2:>8.5f}".format(0, x0, f(x0)))

    xopt = [x0]
    yopt = [f(x0)]

    for i in range(0, epochs):
        x1 = x0 - learning_rate * grad_f(x0)
        x0 = x1
        if disp == True:
            print("{0:02d}\t{1:>8.5f}\t{2:>8.5f}".format(i+1, x0, f(x0)))
        xopt.append(x0)
        yopt.append(f(x0))
        
    return xopt, yopt
```

```python
x_do, y_hat = steepest_descent(f, grad_f, 0.0, learning_rate = 0.25, epochs = 10)

plt.figure(figsize = (10, 5))
plt.plot(x, y, 'k-')
plt.plot(2, 2, 'sk') # a minimum value
plt.plot(x_do, y_hat, 'b*:')
for k, point in enumerate(x_do):
    plt.text(point, f(point), '$x_{}$'.format(k), 
            verticalalignment = 'bottom')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('$x^{2}-4x+6$')
```

```python
x_do, y_hat = steepest_descent(f, grad_f, 0.0, learning_rate = 0.05, epochs = 10)

plt.figure(figsize = (10, 5))
plt.plot(x, y, 'k-')
plt.plot(2, 2, 'sk') # a minimum value
plt.plot(x_do, y_hat, 'b*:')
for k, point in enumerate(x_do):
    plt.text(point, f(point), '$x_{}$'.format(k), 
            verticalalignment = 'bottom')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('$x^{2}-4x+6$')
```

```python
x_do, y_hat = steepest_descent(f, grad_f, 0.0, learning_rate = 1.05, epochs = 10)

plt.figure(figsize = (10, 5))
plt.plot(x, y, 'k-')
plt.plot(2, 2, 'sk') # a minimum value
plt.plot(x_do, y_hat, 'b*:')
for k, point in enumerate(x_do):
    plt.text(point, f(point), '$x_{}$'.format(k), 
            verticalalignment = 'bottom')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('$x^{2}-4x+6$')
```

$\min_{x, y}f(x, y) = 3(x-2)^2 + (y-2)^2$

```python
f = lambda x, y: 3*(x-2)**2 + (y-2)**2 #loss function
```

```python
xmin, xmax, xstep = -8.0, 8.0, .25
ymin, ymax, ystep = -4.0, 8.0, .125

x, y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep),
                  np.arange(ymin, ymax + ystep, ystep))
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm
```

```python
plt.figure(figsize = (25, 10))

ax = plt.axes(projection = '3d', elev = 50, azim = 50)
ax.plot_surface(x, y, f(x, y), norm = LogNorm(),
                rstride = 1, cstride = 1, edgecolor = 'none', 
                alpha = .8, cmap = plt.cm.jet)
minima_ = np.array([[2], [2]])
ax.plot(*minima_, f(*minima_), '*r', markersize = 15)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

plt.show()
```

```python
grad_fx = lambda x, y: 6*(x-2)
grad_fy = lambda x, y: 2*(y-2)
```

```python
dz_dx = grad_fx(x, y)
dz_dy = grad_fy(x, y)

fig, ax = plt.subplots(figsize = (10, 8))
ax.contour(x, y, f(x, y), levels = np.logspace(0, 5, 35), 
           norm = LogNorm(), cmap = plt.cm.jet)
ax.plot(*minima_, '*b', markersize = 18)
ax.quiver(x, y, -dz_dx, -dz_dy, alpha = .7)   #quiver(x, y, u, v): display the velocity vectors as arrows with components (u,v) at the points (x, y) 
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.axis('equal')
```

```python
def steepest_descent_v2(f, grad_fx, grad_fy, init, learning_rate = 1e-1, epochs = 10, disp = True):

    x0 = init
    
    if disp == True:
        print("step    \t  (x, y) \t  f(x, y)")
        print("{0:02d}\t({1:>8.5f}, {2:>8.5f}) : {3:5.4f}".format(0, x0[0], x0[1], f(*x0)))

    xopt = [x0]
    yopt = [f(*x0)]

    for i in range(0, epochs):
        grad = np.array([grad_fx(*x0), grad_fy(*x0)])
        x1 = x0 - learning_rate * grad
        x0 = x1
        if disp == True:
            print("{0:02d}\t({1:>8.5f}, {2:>8.5f}) : {3:5.4f}".format(i+1, x0[0], x0[1], f(*x0)))
        xopt.append(x0)
        yopt.append(f(*x0))
        
    return xopt, yopt
```

```python
x_do, y_hat = steepest_descent_v2(f, grad_fx, grad_fy, np.array([-2., -2.]),
                   learning_rate = 0.25, epochs = 10, disp = True)
```

```python
x_do = np.array(x_do)
x_do = np.array(np.matrix(x_do).T)
```
