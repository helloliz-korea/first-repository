```python
import face_recognition
import os

def get_gropped_face(image_file):
    
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_face = image[a:c,d:b,:]
    
    return cropped_face
```

## 얼굴 영역의 임베딩 추출하기

* me1.jpg에 포함된 얼굴 영역의 임베딩 벡터 구하기  
* 이전 스텝에서 이미지 파일의 얼굴 영역을 가져오는 get_gropped_face() 함수를 활용  
* Face Recognition 패키지에서 제공하는 얼굴 임베딩 벡터를 구하는 함수는 face_recognition.face_encodings()  
* 다음 코드와 같이 사용

```python
image_file = os.path.join(dir_path, 'me1.jpg')
face = get_gropped_face(image_file)   # 얼굴 영역을 구하는 함수(이전 스텝에서 구현)

# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
embedding
```

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    
    for file in file_list:
        
        img_path = os.path.join(dir_path, file)
        face = get_gropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0:
            # 얼굴영역 face가 제대로 detect되지 않으면  len(embedding)==0인 경우가 발생하므로 
            # os.path.splitext(file)[0]에는 이미지파일명에서 확장자를 제거한 이름이 담김
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
        
    return embedding_dict
```

```python
# name1과 name2의 거리를 비교하는 함수를 생성하되, name1은 미리 지정하고,
# name2는 호출시에 인자로 받도록 함
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1

sort_key_func = get_sort_key_func('me2')   
# 이렇게 생성된 함수 sort_key_func는 sort_key_func('me1') 라고 호출할 때,
# me2와 me1 사이의 임베딩 벡터 거리를 계산
```

```python
# 내 사진과 다른 사람 사진에서 추출한 임베딩 벡터들 간의 거리들 중 가장 가까운 거리 5명 추출하는 함수
def get_nearest_face(name, top=5):
    
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))

    for i in range(top+1):
        if i == 0 :   # 첫번째로 나오는 내 이름 제외하기 
            continue
        if sorted_faces[i]:
            print('순위 {} : 이름({}), 거리({})'.format(i, sorted_faces[i][0], sort_key_func(sorted_faces[i][0])))
```
